```python
import pandas as pd

df_2000=pd.read_csv("d1.csv")
```

```python
from sklearn.impute import KNNImputer

# Create a KNNImputer object
imputer = KNNImputer(n_neighbors=5)

# Iterate over the columns and perform imputation
for column in df_2000.columns:
    # Check if the column has missing values
    if df_2000[column].isnull().any():
        # Get the indices of the missing values
        missing_indices = df_2000[column].isnull()

        # Extract the feature matrix X for imputation
        X = df_2000.loc[~missing_indices].values

        # Perform imputation on the feature matrix
        X_imputed = imputer.fit_transform(X)

        # Assign the imputed values back to the DataFrame
        df_2000.loc[~missing_indices] = X_imputed
```

```python
from scipy.cluster.hierarchy import fcluster

# Set the threshold for cutting the dendrogram
threshold = 5 # Adjust this value based on the dendrogram

# Step 1: Determine the number of clusters
clusters = fcluster(linkage_matrix, threshold, criterion='distance')

# Step 2: Get the unique cluster labels and the number of clusters
unique_clusters = np.unique(clusters)
num_clusters = len(unique_clusters)
print("Number of clusters:", num_clusters)

# Step 3: Assign cluster labels to data points
df_scaled['Cluster'] = clusters

# Step 4: Print the clusters and data points within each cluster
for cluster_label in unique_clusters:
    # Select data points belonging to the current cluster
    cluster_data_points = df_scaled[df_scaled['Cluster'] == cluster_label]
    
    # Display the data points of the current cluster
    print(f"Cluster {cluster_label}:")
    display(cluster_data_points)
```
